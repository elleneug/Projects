```
#Writing a transformation that will return the word count of each word present in the file as (key, value) pair
```

```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("distinct")
sc = SparkContext.getOrCreate(conf=conf)
```

```
#read the file with rdd
```

```
rdd = sc.textFile('/FileStore/tables/42_quiz.txt')
rdd.collect()
```

```
rdd2 = rdd.flatMap(lambda x: x.split(' '))
```

```
rdd2.collect()
```

```
rdd3 = rdd2.map(lambda x: (x,1))
rdd3.collect()
```

```
rdd3.reduceByKey(lambda x,y: x+y).collect()
```
