```
#calculate the length of words for a given file
```

```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Read File")
sc = SparkContext.getOrCreate(conf=conf)
```

```
rdd = sc.textFile('/FileStore/tables/doc-3.txt')
```

```
rdd.collect()
```

```
rdd2 = rdd.map(lambda x: x.split(' '))
rdd2.collect()
```

```
def foo(x):
  l = x.split()
  l2 = []
  for s in l:
    l2.append(len(s))
  return l2

rdd3 = rdd.map(foo)
rdd3.collect()
```
