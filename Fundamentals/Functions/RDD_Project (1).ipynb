```
from pyspark import SparkConf, SparkContext
conf = SparkConf().setAppName("Project")
sc = SparkContext.getOrCreate(conf=conf)
```

```
#look at the data and get some insights
```

```
rdd = sc.textFile('/FileStore/tables/StudentData.csv')
```

```
#let's remove headers from further research
```

```
headers = rdd.first()
```

```
rdd = rdd.filter(lambda x: x!= headers)
```

```
#total number of students
```

```
rdd.count()
```

```
# total marks by male and female students
```

```
rdd = rdd.map(lambda x: x.split(','))  
```

```
rdd_mf = rdd.map(lambda x: (x[1], int(x[5])))
```

```
rdd_mf.reduceByKey(lambda x, y: x+y).collect()
```

```
#total passed and failed students
```

```
passed = rdd.filter(lambda x: float(x[5])>50).count()
failed = rdd.filter(lambda x: float(x[5])<=50).count()
```

```
print('N of students that passed the exam:', passed)
print('N of students that failed the exam:', failed)
```

```
#n of total enrollments per course
```

```
rdd_enr = rdd.map(lambda x: (x[3], 1))
```

```
rdd_enr.reduceByKey(lambda x , y : x+y).collect()
```

```
#total marks per course
```

```
rdd_marks = rdd.map(lambda x: (x[3], int(x[5])))
rdd_marks.reduceByKey(lambda x , y : x+y).collect()
```

```
#average marks per course
```

```
rdd_marks_2 = rdd.map(lambda x: (x[3], (int(x[5]), 1)))
```

```
rdd_marks_2.reduceByKey(lambda x,y: (x[0]+y[0], x[1]+y[1])).map(lambda x: (x[0],x[1][0]/x[1][1])).collect()
```

```
#minimum and maximum marks
```

```
minimum = rdd_marks.reduceByKey(lambda x , y : x if x < y else y)
minimum.collect()
```

```
maximum = rdd_marks.reduceByKey(lambda x , y : x if x > y else y)
maximum.collect()
```

```
#average age of male and female students
```

```
rdd_fm_age = rdd.map(lambda x: (x[1], (int(x[0]), 1)))
```

```
rdd_fm_age.reduceByKey(lambda x,y: (x[0]+y[0], x[1]+y[1])).map(lambda x: (x[0],x[1][0]/x[1][1])).collect()
```
